# CineInfini – Démonstration sur une vidéo synthétique

Ce notebook montre comment utiliser CineInfini pour auditer une courte vidéo générée aléatoirement.

1. Installation de CineInfini
2. Audit d’une vidéo synthétique
3. Visualisation du rapport

```python
import cv2
import numpy as np
from pathlib import Path
from cineinfini import audit_video, generate_synthetic_video, CONFIG

# Créer une vidéo synthétique de test
synth_path = Path("synth_demo.mp4")
generate_synthetic_video(str(synth_path), duration_s=5, fps=24, resolution=(320,240), pattern="circle")
print(f"Vidéo créée : {synth_path}")
```

```python
# Lancer l'audit (60 premières secondes)
CONFIG["max_duration_s"] = 60
CONFIG["narrative_coherence"] = False  # accélération sur CPU
report_data, report_dir = audit_video(str(synth_path))
print(f"Rapport sauvegardé dans : {report_dir}")
```

```python
# Afficher le tableau de bord
from IPython.display import display, Markdown
dashboard_path = Path(report_dir) / "dashboard.md"
if dashboard_path.exists():
    with open(dashboard_path, "r") as f:
        display(Markdown(f.read()))
```
